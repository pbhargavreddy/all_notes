# **JSX**
1. JSX allows users to use both html and javascript in a single file. This is readable and easy to write.
2. JSX converts it into HTML elements
    ## RULES
    1. One jsx function can return only one HTML element.  
       ``` 
            Eg. return (
                <h1>Hello</h1>
                <p>Hi</p>
                    )
                    Is invalid ❌```
        Instead use a div box or fragment
    2. Use **classname** not class. Beacuse class is reserved in js
    3. All tags must close  
        ```
            Eg. <img />
                or
                <img></img>```
    4. For javascript inside HTML part use -> **{}**
    5. For HTML inside javascript use -> **()**
# **Componets**
1. A component is a resuabe block or element or box in react.
2. Always use pascal case to name a component.
3. A component can take input arguments too.(Props)
    ```
    Eg:
    <!-- component -->
    function Header(props) {            #Instead of props we can use {name , role} to make it clean and easy to debug 
        return(
            <h1>Hello {props.name}</h1>
        )
    }
    <!-- Using it -->
    function App() {
        return (
            <div>
            <Header name="Bhargav" />
            </div>
            )
    }
    ```

# **UseState**
1. State is used to dynamically change data. Example, when liked a post it instantly shows count increasing.
2. It is used when data need to change, UI needs rerendering after data change
3. Syntax :   
    ```
            const [count, setValue] = useState(initialValue)
    ```
# **UseEffect**
1. It controls when a code block should run
2. Syntax
    ```
            useEffect(()=>{
                //Functiion code
            },[dependency array])
    ```
    ## How to use:
    1. Leave dependency array empty then the function runs every time browser refreshes.
    2. Don't pass the argument it runs every render
    3. Fill it with state variable then it runs every time state changes
