1. When a device is connected to wifi or ethernet, the router allocates an ip for it.   
Nmap can scan all devices connected in that network.
2. The router will expose public ip adress. Router will block all incoming traffic. 
3. ROuter will have public ip adress.
# Public Ip vs Private Ip
1. public ip is visible to every one on the internet. It is unique. It can be static or dynamic. Router will expose the public ip. 
# **NMAP**
1. Nmap is used to scan for devices in a network.
2. It can scan for open ports on a device by sending packets to that ports.
3. It can scan hostnames like google.com,youtube.com
4. Nmap can scan any device it can reach. Not just on local wifi or ethernet.
    ## Same network
    1. Easy.
    2. Needs private ip allocated by wifi router to scan a device.
    ## Any device
    1. Hard.
    2. Need public ip of a device. 
    3. Router blocks all incomming traffic. ROuter internally allocates ip adress to every device and expose a public ip adress to world.
    4. Unless we manually allows incomming traffic others cant scan you using public ip. 
    5. It opens when we allow port forwarding.
