# Installing and encorporating
1. Run **_npm install tailwindcss tailwindcss @tailwindcss/vite_**
2. Navigate to vite.config.js ->   
    import tailwindcss from '@tailwindcss/vite'  
    then add **tailwindcss()** to plugins
3. import in index.csss  -> **@import "tailwindcss";
4. Good to go now
