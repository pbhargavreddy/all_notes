# **Backtracing**
1. for loop backtracking recursion tree. 
    -> It gives all subset of a given set. At every call set is a valid subset.(Subsequence)  
    Tri:
```
    []
    ├── [1]
    │   ├── [1,2]
    │   │   └── [1,2,3]
    │   └── [1,3]
    ├── [2]
    │   └── [2,3]
    └── [3]
```


````python
# Code :
res = []
nums = [1,2,3]
def backtrack(start, path):
        res.append(path.copy())
        for i in range(start, len(nums)):
            path.append(nums[i])   # choose
            backtrack(i+1, path)   # explore
            path.pop()             # undo
backtrack(0,[])
print(*res,sep='\n')
````

2. Subsequences using recursion [Pick , Not pick]
3. Recursion tree
```
        left -> Pick
        right -> Not pick

        lvl->0                     [],0
                                /       \
                            /         \
        lvl->1             [1],1         [],1
                            /  \          / \ 
                        /     \        /   \
        lvl->2          [1,2],2  [1],2  [2],2  [],2

        return Only when idx == len(nums)
        final return res = [[1,2],[1],[2],[]]
```
4. **If you need to return only one result, make the function return boolean. Append the path to res whene true basecase.**
5. **If you need to return count of results, make the function return integer. In function retrun left+right func.**

````python
def printF(res,path,idx,nums):
    if idx == len(nums):
        return res.append(path.copy())
    #Pick
    printF(res,path+[nums[idx]],idx+1,nums)
    #Not pick
    printF(res,path,idx+1,nums)
    return res

nums = [1,2,3]
res = printF([],[],0,nums)
print(*res , sep='\n')
````
