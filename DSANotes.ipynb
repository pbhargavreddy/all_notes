# **Backtracing**
1. for loop backtracking recursion tree. 
    -> It gives all subset of a given set. At every call set is a valid subset.(Subsequence)  
    Tri:
```
    []
    ├── [1]
    │   ├── [1,2]
    │   │   └── [1,2,3]
    │   └── [1,3]
    ├── [2]
    │   └── [2,3]
    └── [3]
```


````python
# Code :
res = []
nums = [1,2,3]
def backtrack(start, path):
        res.append(path.copy())
        for i in range(start, len(nums)):
            path.append(nums[i])   # choose
            backtrack(i+1, path)   # explore
            path.pop()             # undo
backtrack(0,[])
print(*res,sep='\n')
````

2. Subsequences using recursion [Pick , Not pick]
3. Recursion tree
```
        left -> Pick
        right -> Not pick

        lvl->0                     [],0
                                /       \
                            /         \
        lvl->1             [1],1         [],1
                            /  \          / \ 
                        /     \        /   \
        lvl->2          [1,2],2  [1],2  [2],2  [],2

        return Only when idx == len(nums)
        final return res = [[1,2],[1],[2],[]]
```
4. **If you need to return only one result, make the function return boolean. Append the path to res whene true basecase.**
5. **If you need to return count of results, make the function return integer. In function retrun left+right func.**

````python
# Subsequent arrays
def printF(res,path,idx,nums):
    if idx == len(nums):
        return res.append(path.copy())
    #Pick
    printF(res,path+[nums[idx]],idx+1,nums)
    #Not pick
    printF(res,path,idx+1,nums)
    return res

nums = [1,2,3]
res = printF([],[],0,nums)
print(*res , sep='\n')
````

# Merge sort
1. Understanding mergsort gives better grip on recursion and backtracking.
2. And divide and conquer too.
3. Try to write code without slicing the array. Just by using pointers.

````python
#merge sort
def mergeSort(nums):
    def merge(left,right):
        res = []
        i = 0
        j = 0
        while i<len(left) and j< len(right):
            if left[i] < right[j]:
                res.append(left[i])
                i +=1
            else:
                res.append(right[j])
                j+=1
        # Check for elements in left subarray
        while i<len(left):
            res.append(left[i])
            i+=1
        # check for right subarray
        while j<len(right):
            res.append(right[j])
            j+=1
        return res

    def divide(arr):
        if len(arr) <= 1:
            return arr
        mid = len(arr) // 2
        left = divide(arr[:mid])
        right= divide(arr[mid:])
        return merge(left,right)

    return divide(nums)

print(mergeSort([2,1,4,3,5,8,6]))
            
        
````

````python
#Quick sorting
def quickSort(nums,l,r):
    i = l
    j = r
    pivot = l
    while i<j:
        if nums[pivot] < nums[i]:
            while j>i:
                if nums[j] < nums[pivot]:
                    nums[i],nums[j] = nums[j],nums[i]
                j-=1
        i +=1
    nums[i],nums[pivot] = nums[pivot],nums[i]
    return nums

print(quickSort([2,1,4,3,5,8,6],0,6))
````
