# Public key vs private key  (i.e Asymmetric vs symmetric encryption):
1. Let's see them

	## Symmetric
	1. In symmetric encryption receiver and sender will have a secret key. It is used for both encryption and decryption.Examples AES, DES, ChaCHa20, 3DES.
	2. **_UseCase :_** It is used for **session** creation between **browser** and any **website**. Browser gets public key from domain. Then checks certificate and encrypts session key with public key. _**Website decypts**_ with private key.

	## Asymmetric
	1. In asymmetric encryption receiver send a public key to every one and private key to himself. Sender does Encryption with public key. Decryption can be done only with private key receiver have. Both public and private key are mathematically related. Private key can be derived from public key. Examples RSA, Eliptic curve etc.

# Block ciper vs Stream ciper
1. A stream cipher is one that encrypts a digital data stream one bit or one byte at a time.
2. A block cipher is one in which a block of plaintext is treated as a whole and used to produce a ciphertext block of equal length. Typically, a block size of 64 or 128 bits is used.
